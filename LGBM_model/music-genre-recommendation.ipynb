```python
import numpy as np
import pandas as pd
import pickle
import joblib
import sklearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
scaler = joblib.load('../input/music-dataset/standard_scaler.save')
model = pickle.load(open('../input/music-dataset/opt_lgbm_model.pkl', 'rb'))
```

```python
df = pd.read_csv('../input/gtzan-dataset-music-genre-classification/Data/features_30_sec.csv')
df
```

```python
labels = df[['label']]
scaled_df = scaler.transform(df.iloc[:, 1:-1])
scaled_df
```

```python
similarity = cosine_similarity(scaled_df)
df_similarity = pd.DataFrame(similarity)
df_similarity.columns = df['filename']
df_similarity = df_similarity.set_index(df['filename'])
df_similarity
```

```python
def recommend_similar_songs(song_name):
    recommendations = df_similarity[song_name].sort_values(ascending = False)
    recommendations = recommendations.drop(song_name)
    return recommendations.head(5)

print('=============For Blues=============')
print(recommend_similar_songs('blues.00015.wav'))
print('=============For Rock=============')
print(recommend_similar_songs('rock.00028.wav'))
print('=============For Pop=============')
print(recommend_similar_songs('pop.00031.wav'))
print('=============For Jazz=============')
print(recommend_similar_songs('jazz.00011.wav'))
```

```python
def make_recommendation_df(song_name):
    recommendations = recommend_similar_songs(song_name)
    recommendations = pd.DataFrame(recommendations).to_csv('rec.csv') 
    rec_songs = pd.read_csv('./rec.csv')
    rec_songs.rename(columns = {song_name: 'cosine_sim_score'}, inplace = True)
    return rec_songs

make_recommendation_df('jazz.00011.wav')
```
